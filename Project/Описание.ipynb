# Цель работы.

---


Используя веб парсинг, сформировать базу данных по интересующим направлениям с ресурса coursera.org, содержащую информацию по курсам. На основе полученных данных построить рекомендательную модель.


---
# План работы.


1.   Собрать данные по курсам и сохранить в базу данных в формате csv, содержащую информацию:
*   Название курса
*   Описание курса
*   Оценка курса
*   Рейтинг курса
*   Ссылка на курс
*   Направление курса

При поиске данных учитываются программы с типом курс и только на английском языке.

2.   Обработать полученные данные:

*   Отфильтровать курсы, не имеющие оценок и рейтинга;
*   Для построения рекомендательной модели разобьем описание курсов на токены (слова), предварительно очистив данное поле от ссылок;
*   Перевести слова в нижний регистр;
*   Избавится от стоп-слов;
*   Произвести лемматизацию слов;
*   Избавится от символов, не являющихся буквенными.

3.  Произвести анализ данных:

*   Сформировать гистограмму средних значений оценок и рейтинга по каждому направлению;
*   Оценить сбалансированность числа курсов в зависимости от направлений;
*   Оценить сбалансированность числа слов в зависимости от направлений;
*   Выполнить кодировку целевого поля - направления;
*   Сформировать статистику важности слов на основе алгоритма TF-IDF;
*   Выполнить обучение модели на основе тестового набора данных;
*   Сформировать отчет и сделать выводы о качестве модели классификации;
*   На основе введенных данных пользователем сформировать рекомендации (до 10) по курсам, учитывая релевантность направления полученную из модели, а так же количества оценок и рейтинга курсов.
